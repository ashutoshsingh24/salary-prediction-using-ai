```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('salary.csv')
print(df.head())
```

```python
df = pd.read_csv('salary.csv')
print(df.head())
```

```python
x=df[['Years of Experience']]
y=df['Salary']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train_scaled, Y_train)
```

```python
Y_pred = knn.predict(X_test_scaled)

mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"Predicted Salaries: {Y_pred}")
print(f"Mean Squared Error: {mse}")
print(f"RÂ² Score: {r2}")
```

```python
years_exp = float(input("Enter years of experience: "))

input_scaled = scaler.transform([[years_exp]])
predicted_salary = knn.predict(input_scaled)

print(f"Predicted Salary for {years_exp} years of experience: â‚¹{predicted_salary[0]:,.2f}")
```
